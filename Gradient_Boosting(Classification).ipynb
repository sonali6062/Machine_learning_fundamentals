```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
data=load_iris()
```

```python
data
```

```python
# Convert the data to a pandas DataFrame
df = pd.DataFrame(data.data, columns=data.feature_names)

# Add the target variable to the DataFrame
df['target'] = data.target

# Display the first few rows and the columns
display(df.head())
display(df.columns)
```

```python
## independent feature-x,target feature-y
x,y=data.data,data.target
print(x)
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
model=GradientBoostingClassifier()
model.fit(x_train,y_train)
```

Perform Hyperparmeter tuning vis GridSearchCV and compare it with without tuning model

MODEL PREDICTION


```python
y_pred=model.predict(x_test)
y_pred
```

MODEL EVALUATION

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy",accuracy)
print('-'*30)
print("Confusion Matrix:")
print(confusion_matrix(y_test,y_pred))
print('-'*30)
print("Classification Report:")
print(classification_report(y_test,y_pred))
```

Accuracy is 1 means 100 percent. It is a clear indication that the model is overfitted(this might occur due to less record size or data leackage)

Data Modelling:GradientBoostingRegressor

```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error
```

```python
housing=fetch_california_housing()
x=housing.data
y=housing.target
```

Data Splitting: Split the data into independent features and target values.

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(y_train)
```

Data Modelling

```python
model_reg=GradientBoostingRegressor()
model_reg.fit(x_train,y_train)
```

```python
# Make predictions on the test set using the regression model
y_pred_reg = model_reg.predict(x_test)

y_pred_reg
```

```python
# Evaluate the regression model using mean squared error
mse = mean_squared_error(y_test, y_pred_reg)
print(f"Mean Squared Error: {mse}")
```

Error is very low.
lower the value of MSE higher the accuracy of the model

```python
from sklearn.metrics import r2_score

# Calculate R-squared score
r2 = r2_score(y_test, y_pred_reg)
print(f"R-squared Score: {r2}")
```

# Hyperparameter tuning on a Gradient Boosting Classifier

```python
param_grid = {
    'n_estimators': [100, 200],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 4, 5]
}
```

## Initialize gridsearchcv




```python
from sklearn.model_selection import GridSearchCV

grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, scoring='accuracy')
```

```python
# Re-split the Iris data for classification
x_train_clf, x_test_clf, y_train_clf, y_test_clf = train_test_split(data.data, data.target, test_size=0.2, random_state=42)

# Fit GridSearchCV to the Iris training data
grid_search.fit(x_train_clf, y_train_clf)
```

# Get best model
Extract the best performing model from the `GridSearchCV` results.


```python
best_model = grid_search.best_estimator_
best_model
```

# Make predictions
Use the best model to make predictions on the test data (x_test).


```python
y_pred_tuned = best_model.predict(x_test_clf)
```

```python
accuracy_tuned = accuracy_score(y_test_clf, y_pred_tuned)
print(f"Accuracy of the tuned model: {accuracy_tuned}")
```

```python
print("Best hyperparameters found:")
print(grid_search.best_params_)
print("-" * 30)
print("Accuracy of the tuned model:")
print(accuracy_tuned)
```
