```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

LOAD THE DATASET

```python
data=pd.read_csv("diabetes_data.csv")
data.head()
```

SPLIT THE DATA INTO INPUT AND TARGET FEATURES

```python
#Split the data into input feature and teh target outcome
x=data.drop(columns=['diabetes'],axis=1)
y=data['diabetes']
```

SPLIT THE ENTIRE DATA INTO TEO SUBSETS:
SUBSET1(TRAIN SET)
SUBSET2(VALIDATION SET/HOLD OUT SET)

```python
from sklearn.model_selection import train_test_split
train,val_train,test,val_test=train_test_split(x,y,test_size=0.5,random_state=355)
```

SPLIT THE TRAIN SET FURTHER INTO TRAIN AND TEST SET(80:20)

```python
  x_train,x_test,y_train,y_test=train_test_split(train,test,test_size=0.2)
```

MODEL1-implementation of KNeighborsClassifier


```python
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
knn.score(x_test,y_test)
```

MODEL2-Implmentation of SVC

```python
svc=SVC()
svc.fit(x_train,y_train)
svc.score(x_test,y_test)
```

Prediction of the model in the validation set

```python
predict_val1=knn.predict(val_train)
predict_val2=svc.predict(val_train)
```

```python
predict_val1
```

```python
predict_val2
```

COLUMN STACK-combine both models

```python
predict_val=np.column_stack((predict_val1,predict_val2))
```

```python
predict_val#[x,y]x-knn,y-svc
```

```python
predict_test1=knn.predict(x_test)
predict_test2=svc.predict(x_test)
predict_test=np.column_stack((predict_test1,predict_test2))
```

STACKING: Input features are the predictions of the model M1(KNN) and model2 M2(SVC) stored insid ethe predict_val(input  geatures)for the RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
rand_clf=RandomForestClassifier()
rand_clf.fit(predict_val,val_test)#predict_val-model training using both SVC and knn . It is not the original data
```

```python
rand_clf.score(predict_test,y_test)
```

Hyperparameter tuning-GridSeachCV

```python
data.columns
```

```python
from sklearn.model_selection import GridSearchCV
grid_param={
    "n_estimators":[90,100,115],
    "criterion":['gini','entropy'],
    "min_samples_leaf":[1,2,3,4,5],
    "min_samples_split":[4,5,6,7,8],
    "max_features":['sqrt', 'log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
grid_search=GridSearchCV(estimator=rand_clf,param_grid=grid_param,cv=5,n_jobs=-1,verbose=3)
```

```python
grid_search.fit(predict_val,val_test)
```

Fitting 5 folds for each of 300 candidates, totalling 1500 fits

```python
grid_search.best_params_
```

```python
rand_clf_hyperparameter_tuning=RandomForestClassifier(criterion='gini',max_features='sqrt',min_samples_leaf=1,min_samples_split=4,n_estimators=90)
rand_clf_hyperparameter_tuning.fit(predict_val,val_test)
```

```python
rand_clf_hyperparameter_tuning.score(predict_test,y_test)
```
