# STEP 2- covariance matrix of given dataframe

```python
 #APPROACH1
 c1=dataframe.x1
 c2=dataframe.x2
 c3=dataframe.x3
```

```python
np.cov(dataframe)
```

```python
#APPROACH2
covarianceMatrix=dataframe.T @ dataframe/4
covarianceMatrix
#@-for matrix multiplication/(n-1)
```

```python
#Approch3
np.sum(c1*c2*c3)/4
```

# Step3- Evaluate the eigen value and eigen vector of correlation matrix

```python
eigenValue,eigenVectors=np.linalg.eig(covarianceMatrix)
```

```python
eigenValue
```

```python
eigenVectors
```

```python
eigenVectors[:,0]
```

```python
dataframe
```

```python
PC1=dataframe @ eigenVectors[:,0]
PC1
#most of the features from dimentions are extracted
```

```python
PC2=dataframe @ eigenVectors[:,1]
PC2
```
